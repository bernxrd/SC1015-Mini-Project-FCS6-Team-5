```python
import pandas as pd

# Load the Excel file to see its content and structure
file_path = r"C:\Users\raysh\OneDrive - Nanyang Technological University\Year 1 Sem 2\SC1015 Intro to Data Sci and AI\Mini Project\Nvidia Data\Updated_GPU_Dataset279_Nvidia.xlsx"
gpu_data = pd.read_excel(file_path)

# Display the first few rows of the dataframe to understand its structure
gpu_data.head()
```

```python

from bs4 import BeautifulSoup
import requests
import time 
import random
```

```python
def classify_gen(arch):
    
    gpu_details = {
        "Arbitrary Gen": "Unknown",
        "Gen Year": "Unknown"
    }
    
    if arch == "Tesla":
        gpu_details = {
            "Abritary Gen": "1",
            "Gen Year": "2006",
        }
    if arch == "Tesla 2.0":
        gpu_details = {
            "Abritary Gen": "2",
            "Gen Year": "2007",
        }
    if arch == "Fermi":
        gpu_details = {
            "Abritary Gen": "3",
            "Gen Year": "2010",
        }
    if arch == "Fermi 2.0":
        gpu_details = {
            "Abritary Gen": "4",
            "Gen Year": "2011",
        }
    if arch == "Kepler":
        gpu_details = {
            "Abritary Gen": "5",
            "Gen Year": "2012",
        }
    
    if arch == "Kepler 2.0":
        gpu_details = {
            "Abritary Gen": "6",
            "Gen Year": "2013",
        }
    
    if arch == "Maxwell":
        gpu_details = {
            "Abritary Gen": "7",
            "Gen Year": "2014",
        }
    if arch == "Maxwell 2.0":
        gpu_details = {
            "Abritary Gen": "8",
            "Gen Year": "2015",
        }
    if arch == "Pascal":
        gpu_details = {
            "Abritary Gen": "9",
            "Gen Year": "2016",
        }
    if arch == "Volta":
        gpu_details = {
            "Abritary Gen": "10",
            "Gen Year": "2017",
        }
    if arch == "Turing":
        gpu_details = {
            "Abritary Gen": "11",
            "Gen Year": "2018",
        }
    if arch == "Ampere":
        gpu_details = {
            "Abritary Gen": "12",
            "Gen Year": "2020",
        }
    if arch == "Ada Lovelace":
        gpu_details = {
            "Abritary Gen": "13",
            "Gen Year": "2022",
        }
        
    return gpu_details
```

```python

iteration = 0
start_index = 0
```

```python

# Iterate over the DataFrame and scrape details for each GPU
print(f'the start iteration is {iteration}')
print(f'the start index is {start_index}')
for index, row in gpu_data.loc[start_index:].iterrows():
    iteration += 1
    print(f"the iteration is {iteration}")
    gpu_details = classify_gen(row['Architecture'])
    for key, value in  gpu_details.items():
        gpu_data.at[index, key] = value
  



# Save the updated DataFrame to a new Excel file once all done
updated_file_path = f'Updated_GPU_Dataset{iteration}_Nvidia_gen.xlsx'
gpu_data.to_excel(updated_file_path, index=False)
print(f"Updated dataset saved to {updated_file_path}")
```
